# Observation 1

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,palette='BuGn',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

# Observation 2
## Highest no. of buyers from California and NY

```python
df.hist(figsize=(10,10), bins = 50)
plt.show()
```

# Observation 3
## Most customers tend to buy quantity of 2 and 3
## Discount max. 0 to 20 %


```python
plt.figure(figsize=(15,10))
df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

# Profit vs Discount

```python
figure, axis = plt.subplots(figsize=(15,10))
axis.scatter(df['Sales'],df['Profit'])
axis.set_xlabel('Sales')
axis.set_xlabel('Profit')
plt.show()
```

```python
sns.lineplot(x = 'Discount', y = 'Profit', label = 'Profit' , data = df)
plt.legend()
plt.show()
```

# Observation 4
## No correlation between profit and discount

# Profit vs Quantity

```python
sns.lineplot(x = 'Quantity', y = 'Profit', label = 'Profit', data = df)
plt.legend()
plt.show()
```

# Observation 5
## Profit and sales max. in consumer segment and min. in Home Office segment


```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='Accent')
plt.xlabel('Region',fontsize=15)
plt.show()
```

# Observation 6
## Segment wise sales similar in every region


```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['green','orange'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# Observation 7
## Profit and sales are max. in west region and min. in south region


```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

# Observation 8
## High profit in California and NY
## Loss in Texas, Pennsylvania, Ohio



```python
states = df['State'].value_counts().nlargest(10)
print(states)
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','green'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# Observation 9
## Technology and Office Supplies have high profit
## Furniture has less profit



```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightgreen'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

# In conclusion, 
## Phones sub-category have high sales.
## Chairs have high sales but less profit compared to phones.
## Tables and Bookmarks sub-categories face huge loss.
